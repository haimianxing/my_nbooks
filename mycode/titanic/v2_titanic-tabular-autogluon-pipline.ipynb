```python
df = pd.read_csv('./df.csv')
df_test = pd.read_csv('./df_test.csv')
df
```

```python
# 数据清洗
df['Age'].fillna(df['Age'].median(), inplace=True)  # 年龄中位数填充
df['Fare'].fillna(df['Fare'].median(), inplace=True)  # 票价中位数填充
df.dropna(subset=['Embarked'], inplace=True)  # 删除Embarked缺失的行
df=df.drop_duplicates()
df
```

```python
df 
```

```python
import numpy as np

# 特征创建
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)  # 从名字中提取头衔
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1


# 社会经济状态特征
# df['SocioEconomicStatus'] = df.apply(lambda row: 'High' if row['Pclass'] == 1 else (
#     'Middle' if row['Pclass'] == 2 else 'Low'), axis=1)
# df['SocioEconomicStatus'] = df['SocioEconomicStatus'] + '_' + df['Title'] + '_' + pd.cut(df['Fare'], bins=[0, 8, 15, float('inf')],
#                                                                                      labels=['LowFare', 'MidFare', 'HighFare'])

# 年龄分段特征
age_bins = [0, 12, 18, 60, np.inf]
age_labels = ['Child', 'Teenager', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# 家庭结构特征
df['HasChildren'] = df['Parch'] > 0
df['HasSpouse'] = df['SibSp'] > 0

# 姓名长度特征
df['NameLength'] = df['Name'].apply(lambda x: len(x))

# 客舱区域特征
df['CabinArea'] = df['Cabin'].astype(str).apply(lambda x: x[0] if pd.notnull(x) else 'Unknown')

# 登船时间特征（这里简化处理，实际情况可能需要更复杂的解析）
embark_time_mapping = {'C': 'Early', 'Q': 'Mid', 'S': 'Late'}
df['EmbarkTime'] = df['Embarked'].map(embark_time_mapping)

# 票价类别特征
df['FareCategory'] = pd.qcut(df['Fare'], q=4, labels=['Very_Low', 'Low', 'High', 'Very_High'])

# 生存网络特征（此处简化处理，实际分析可能涉及复杂的社会网络构建）
# df['SameLastName'] = df.groupby('Name').transform(lambda x: x['Name'].duplicated().astype(int))

# 打印更新后的DataFrame的前几行以查看新增特征
df
```

```python
#频率编码
from feature_engine.encoding import CountFrequencyEncoder
cf = CountFrequencyEncoder(encoding_method='count',missing_values='ignore')
cf.fit(df_solve_mi)
df_solve_mi_cf=cf.transform(df_solve_mi)

df_solve_mi_cf
```

# 分类空白插值

```python
#分类空白插值
from feature_engine.imputation import MeanMedianImputer
mmi = MeanMedianImputer(imputation_method='median')
mmi.fit(df_solve_mi_cf)
df_solve_mi_cf_mmi = mmi.transform(df_solve_mi_cf)

df_solve_mi_cf_mmi
```
